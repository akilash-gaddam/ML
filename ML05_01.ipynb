```python
mat_score = df['math_score'].to_numpy()

sci_score=df['science_score'].to_numpy()

eng_score=df['english_score'].to_numpy()
overall_score=df['overall_score'].to_numpy()
print(mat_score)
print(np.mean(mat_score))
print(np.median(mat_score))
print(np.std(mat_score))
print(sci_score)
print(np.mean(sci_score))
print(np.median(sci_score))
print(np.std(sci_score))
print(eng_score)
print(np.mean(eng_score))
print(np.median(eng_score))
print(np.std(eng_score))
print(overall_score)
print(np.mean(overall_score))
print(np.median(overall_score))
print(np.std(overall_score))
```

```python
print(df['overall_score'].max())
print(df['math_score'].max())
print(df['english_score'].max())
```

```python
plt.hist(df['math_score'],bins=5,color='red',edgecolor='black')
plt.title('math_performance')
plt.xlabel('math_score')
plt.ylabel('frequency')
plt.show()
```

```python
avg_score=df[['math_score','science_score','english_score']].mean()
print(avg_score)
avg_score.plot(kind='bar',color=['red','green','blue'],edgecolor='black')
plt.title('average_score')
plt.xlabel('subject')
plt.ylabel('average_score')
plt.show()
```

```python
plt.scatter(df['math_score'],df['english_score'],color='red',edgecolor='black')
plt.title('math_vs_english')
plt.xlabel('math_score')
plt.ylabel('english_score')
plt.show()
```

```python
gender_avg_scores = df.groupby('gender')[['math_score', 'science_score', 'english_score', 'overall_score']].mean()
print(gender_avg_scores)
```

```python
parent_education_avg_scores = df.groupby('parent_education')[['math_score', 'science_score', 'english_score', 'overall_score']].mean()
print(parent_education_avg_scores)
```

```python
gender_avg_scores.plot(kind='bar', figsize=(10, 6))
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.legend(title='Score Type')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
